```python
from spacy.lemmatizer import Lemmatizer
from spacy.lang.id import LOOKUP

lemmatizer = Lemmatizer(lookup=LOOKUP)
lemmas = lemmatizer(u'bepergian', u'NOUN')
print(lemmas)
```

```python
with open('hmmscratch/data/test-data.tsv', 'r') as f:
    for word in f.readlines():
        word = word.strip('\n')
        lemmas = lemmatizer(unicode(word), u'NOUN')
        print(lemmas)
f.close()
```

```python
with open('hmmscratch/emission.txt', 'r') as f:
    for word in f:
        word = word.split(None,1)[0]
        lemmas = lemmatizer(unicode(word), u'NOUN')
        print(lemmas)
f.close()
```
